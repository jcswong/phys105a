#### Using python (jupyter notebooks), mathematica, or Wolfram alpha, plot the corresponding truncated Fourier series up to order N, i.e

#### $$f_N (x) = \sum^{n=N}_{n=-N} c_ne^{-inx}$$

#### for N = 2,5,10. Include a screenshot of your code plus resulting plots.

```python
sym.init_printing()
x= sym.symbols('x')

f_1 = sym.sin(x)
f_2 = sym.Piecewise((0, x<=0), (f_1, x<pi))
sym_plot.plot(f_2, (x,-pi,pi))
```

```python
s = fourier_series(f_2, (x, -pi, pi))
s1 = s.truncate(n = 2) # for n = 2
s2 = s.truncate(n = 5) # N = 5
s3 = s.truncate(n = 10) # N = 10

p = plot(f_2, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
p[0].line_color = (0, 0, 0)
p[0].label = 'f(x)'
p[1].line_color = (0.3, 0.3, 0.3)
p[1].label = 'n=2'
p[2].line_color = (0.9, 0.2, 0.3)
p[2].label = 'n=5'
p[3].line_color = (0.3, 0.5, 0.9)
p[3].label = 'n=10'

print("N = 2  " + str(s1) + "\nN = 5  " + str(s2) + "\nN = 10  " + str(s3))
p.show()
```
