```python
sym.init_printing()
t= sym.symbols('t')

a = 1

f_1 = t
f_2 = sym.Piecewise((0, t<=0),(0, t>=a), (f_1, t<=a))
sym_plot.plot(f_2, (t, -a-a/2, a+a/2))
```

```python
s = fourier_series(f_2, (t, 0, a))
s1 = s.truncate(n = 3) # for n = 3
s2 = s.truncate(n = 30) # n = 30

p = plot(f_2, s1, s2, (t, -, a+a/2), show=False, legend=True)
```

```python
p[0].line_color = (0, 0, 0)
p[0].label = 'f(x)'
p[1].line_color = (0.3, 0.3, 0.3)
p[1].label = 'n=3'
p[2].line_color = (0.9, 0.2, 0.3)
p[2].label = 'n=30'

print("N = 3  " + str(s1) + "\nN = 30  " + str(s2))
p.show()
```
